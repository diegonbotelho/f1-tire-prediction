```python
import fastf1
import pandas as pd

# Enable cache to speed up subsequent data loading
fastf1.Cache.enable_cache('./cache')

# Load session data (Example: 2024 Bahrain GP - Race)
session = fastf1.get_session(2024, 'Bahrain', 'R')
session.load()

# Extract lap data into a DataFrame
laps_df = session.laps

# Save lap data to CSV
laps_df.to_csv('2024_bahrain_race_laps.csv', index=False)
```
